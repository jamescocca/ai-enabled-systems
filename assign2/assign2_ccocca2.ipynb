Image Preprocessing

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# Read in the 4 bike images provided for Assignment 2
bike1 = cv2.imread('images/BikeShot1.jpg')
bike2 = cv2.imread('images/BikeShot2.jpg')
bike3 = cv2.imread('images/BikeShot3.jpg')
bike4 = cv2.imread('images/BikeShot4.jpq.png')
plt.imshow(bike1)
```

```python
# Resize all bike images to the standard YOLO input resolution
yolo_dim = (416, 416)
bike1_resized = cv2.resize(bike1, yolo_dim, interpolation = cv2.INTER_AREA)
bike2_resized = cv2.resize(bike2, yolo_dim, interpolation = cv2.INTER_AREA)
bike3_resized = cv2.resize(bike3, yolo_dim, interpolation = cv2.INTER_AREA)
bike4_resized = cv2.resize(bike4, yolo_dim, interpolation = cv2.INTER_AREA)
plt.imshow(bike1_resized)
```

```python
# Normalize the resized bike images
zeros = np.zeros(yolo_dim)
bike1_normalized = cv2.normalize(bike1_resized, zeros, 0, 255, cv2.NORM_MINMAX)
bike2_normalized = cv2.normalize(bike2_resized, zeros, 0, 255, cv2.NORM_MINMAX)
bike3_normalized = cv2.normalize(bike3_resized, zeros, 0, 255, cv2.NORM_MINMAX)
bike4_normalized = cv2.normalize(bike4_resized, zeros, 0, 255, cv2.NORM_MINMAX)
plt.imshow(bike1_normalized)
```
