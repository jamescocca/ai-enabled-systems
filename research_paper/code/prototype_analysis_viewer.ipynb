```python
import random
import os
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
prototypes_dir = 'most_activated_prototypes/'
path_to_orig_patch = (prototypes_dir + 'most_highly_activated_patch_in_original_img_by_top-1_prototype.png')
path_to_prototype = (prototypes_dir + 'top-1_activated_prototype_in_original_pimg.png')
```

```python
def show_input_activation(selection, gender_and_skin_tone):
    print("Most highly activated section of input image:")
    path = './prototype_analysis/' + gender_and_skin_tone + '/' + selection + '/' + path_to_orig_patch
    orig = img.imread(path)
    plt.imshow(orig)
```

```python
def show_proto_activation(selection, gender_and_skin_tone):
    print("Most highly activated prototype image section for above input:")
    path = './prototype_analysis/' + gender_and_skin_tone + '/' + selection + '/' + path_to_prototype
    orig = img.imread(path)
    plt.imshow(orig)
```

Light-skinned male sample

```python
folders = []
for folder in os.listdir('./prototype_analysis/light_skinned_male/'):
    if folder != '.DS_Store':
        folders.append(folder)
selection = folders[random.randrange(len(folders))]
show_input_activation(selection, 'light_skinned_male')
```

```python
show_proto_activation(selection, 'light_skinned_male')
```
